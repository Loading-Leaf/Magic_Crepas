```python
import cv2
from rembg import remove
from PIL import Image
import numpy as np

def remove_background(input_image_path, output_image_path):
    # 背景を削除
    try:
        input_image = Image.open(input_image_path)
    except IOError:
        print(f"Error: Cannot open {input_image_path}")
        return

    output_image = remove(input_image)
    output_image.save(output_image_path)

def apply_mask_to_background(masked_image_path):
    # RGBA画像を読み込み
    rgba_image = cv2.imread(masked_image_path, cv2.IMREAD_UNCHANGED)
    if rgba_image is None:
        print(f"Error: Cannot open {masked_image_path}")
        return

    # アルファチャネルをマスクとして使用
    alpha_channel = rgba_image[:, :, 3]

    # 白い背景画像を作成
    background = np.ones_like(rgba_image, dtype=np.uint8) * 255
    # マスクを適用
    background_masked = cv2.bitwise_and(background, background, mask=alpha_channel)
    return background_masked

def main():
    input_path = 'S__124059653.jpg'
    output_path = './output.png'

    # 背景除去処理
    remove_background(input_path, output_path)

    # マスク適用処理
    masked_image = apply_mask_to_background(output_path)
    if masked_image is not None:
        cv2.imwrite('masked_image.jpg', masked_image)

if __name__ == "__main__":
    main()

```

画風変換アルゴリズム

```python
import os
import tensorflow as tf
# Load compressed models from tensorflow_hub
os.environ['TFHUB_MODEL_LOAD_FORMAT'] = 'COMPRESSED'
```

```python
import IPython.display as display

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (12, 12)
mpl.rcParams['axes.grid'] = False

import numpy as np
import PIL.Image
import time
import functools
```

```python
def tensor_to_image(tensor):
  tensor = tensor*255
  tensor = np.array(tensor, dtype=np.uint8)
  if np.ndim(tensor)>3:
    assert tensor.shape[0] == 1
    tensor = tensor[0]
  return PIL.Image.fromarray(tensor)
```

```python
content_path = 'S__124059653.jpg'
style_path = tf.keras.utils.get_file('kandinsky5.jpg','https://storage.googleapis.com/download.tensorflow.org/example_images/Vassily_Kandinsky%2C_1913_-_Composition_7.jpg')
```

```python
def load_img(path_to_img):
  max_dim = 512
  img = tf.io.read_file(path_to_img)
  img = tf.image.decode_image(img, channels=3)
  img = tf.image.convert_image_dtype(img, tf.float32)

  shape = tf.cast(tf.shape(img)[:-1], tf.float32)
  long_dim = max(shape)
  scale = max_dim / long_dim

  new_shape = tf.cast(shape * scale, tf.int32)

  img = tf.image.resize(img, new_shape)
  img = img[tf.newaxis, :]
  return img
```

```python
def imshow(image, title=None):
  if len(image.shape) > 3:
    image = tf.squeeze(image, axis=0)

  plt.imshow(image)
  if title:
    plt.title(title)
```

```python
content_image = load_img(content_path)
style_image = load_img(style_path)

plt.subplot(1, 2, 1)
imshow(content_image, 'Content Image')

plt.subplot(1, 2, 2)
imshow(style_image, 'Style Image')
```

```python
import tensorflow_hub as hub
hub_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
stylized_image = hub_model(tf.constant(content_image), tf.constant(style_image))[0]
tensor_to_image(stylized_image)
output_style = tensor_to_image(stylized_image)
```

```python
output_style.save("output_style.png")
```

合成

```python
masked_image = cv2.imread('masked_image.jpg')
output_style = cv2.imread('output_style.png')
output_style = cv2.cvtColor(output_style, cv2.COLOR_BGR2RGB)
masked_image = cv2.resize(masked_image, (output_style.shape[1], output_style.shape[0])) #横, 縦
print(masked_image.shape)
print(output_style.shape)
```

```python
dst = cv2.bitwise_and(output_style, masked_image)
plt.subplot(1, 2, 1)
imshow(dst, 'synthesis')
```

```python
masked_background = cv2.bitwise_not(masked_image)
input = cv2.imread('S__124059653.jpg', cv2.IMREAD_UNCHANGED)
input = cv2.cvtColor(input, cv2.COLOR_BGR2RGB)
input = cv2.resize(input, (masked_background.shape[1], masked_background.shape[0])) #横, 縦
dst2 = cv2.bitwise_and(input, masked_background)
plt.subplot(1, 2, 1)
imshow(dst2, 'synthesis2')
```

```python
result = cv2.bitwise_or(dst, dst2)
plt.subplot(1, 2, 1)
imshow(result, 'Result')
```
